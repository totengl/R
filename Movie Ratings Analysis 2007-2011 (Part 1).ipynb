```r
setwd("C:\\Users\\Phoenix_Steve\\Desktop\\R programming A-Z\\Section 6")
movies<- read.csv("P2-Movie-Ratings.csv")
library(ggplot2)
suppressWarnings(suppressMessages(library(ggplot2)))


#head(movies). Looking atfirst six rows

colnames(movies)<- c("Film", "Genre","CriticRating","AudienceRating", "BudgetMillions","Year")

#colnames(movies) Checking col names

#factor(movies$Year)

movies$Year<-factor(movies$Year) #creating levels in the year column

#str(movies) check structure of df

o<-ggplot(data=movies, aes(x=BudgetMillions))
h<- o + geom_histogram(binwidth=10, aes(fill=Genre), colour="Black")
windowsFonts(TimesNewRoman=windowsFont("TT Times New Roman"))
#axeslabels
h + xlab("Money Axis") + ylab("Number of Movies") +
  ggtitle("Movie Budget Distribution")+
  theme(axis.title.x=element_text(colour="DarkGreen", size=30),
        axis.title.y=element_text(colour="Red", size=30),
        axis.text.x=element_text(size=20),
        axis.text.y=element_text(size=20),
        
        legend.title=element_text(size=30),
        legend.text=element_text(size=20),
        legend.position=c(1,1),
        legend.justification=c(1,1),
        
        plot.title=element_text(colour="DarkBlue",
                                size=35,
                                family = "TimesNewRoman"))
```

**Number of movies per Genre released according to budget in miilions in the USA. Data based on movies released from 2007-2011.**

```r
q<-ggplot(data=movies, aes(x=CriticRating,y=AudienceRating,
                              colour=Genre, size=BudgetMillions))

q+ geom_point()   
```

**Audience Rating vs Critic Rating. Where colours show the genre and the size of the point shows the budget in millions. If we were to draw a line x=y on to this graph. It is clear to see that for most movies the audience rating tends to be higher than the critics i.e the audience are more forgiving of movies than critics.**

```r
q+ geom_point(aes(size=CriticRating))
```

**Overriding the size=budget aesthetic. Size now represents the critic rating so as we move right the size of the points increases because the critic rating increases.**

```r
q+ geom_point(aes(colour=BudgetMillions))
```

**Overiding the colour aethetics. Now  the colour represents the budget. The lighter the colour the more expensive the movie**

```r
q+ geom_point(aes(x=BudgetMillions)) + xlab("Budget Millions $$")
```

**Action movies tend to have bigger budgets.**

```r
u<- ggplot(data=movies, aes(x=CriticRating, y=AudienceRating,colour=Genre))

u+ geom_point() + geom_smooth(fill=NA)
```

**Looking at the trend lines. Audience rating doent vary as widely as Critic Rating according to genre. Audience highly rated Romances tend to have a low rating by critics.**

```r
u<- ggplot(data=movies, aes(x=Genre, y=AudienceRating,colour=Genre))

u+geom_boxplot()
```

**Romances, Thrillers, and Dramas tend to be the most highly rated by audiences. Horrors are usually the least highly rated.**

```r
u+geom_boxplot(size=1.2) + geom_jitter()
```

**Same plot as above but showing the actual distribution of ratings.**

```r
u+geom_jitter()+ geom_boxplot(aes(y=CriticRating), size=1.2,alpha=0.5) +
  ylab("CriticRating")
```

**Another plot showing how variable the critic rating tends to be. Thrillers tend to be the most highly rated among critics**

```r
v<- ggplot(data=movies, aes(x=BudgetMillions))
v+geom_histogram(binwidth=10,aes(fill=Genre),colour="Black")+
  facet_grid(Genre~.,scales="free")
```

**Illustrating the use of facet grids.**

```r
w<- ggplot(data=movies, aes(x=CriticRating, y=AudienceRating, colour=Genre))
w + geom_point(aes(size=BudgetMillions)) + facet_grid(Genre~Year)+ geom_smooth() + coord_cartesian(ylim=(c(0,100)))
```

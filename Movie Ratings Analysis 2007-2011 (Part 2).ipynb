```r
setwd("C:\\Users\\Phoenix_Steve\\Desktop\\R programming A-Z\\Section 6")
library(ggplot2)
library(repr)
suppressWarnings(suppressMessages(library(ggplot2)))
mov<- read.csv("Section6-Homework-Data.csv")
colnames(mov)[18]<-"GrosspercentUS"
colnames(mov)[8]<-"Budgetmillions"


mov2<-mov[(mov$Genre=="action"|mov$Genre=="adventure"| 
            mov$Genre=="animation"|mov$Genre=="comedy"|
            mov$Genre=="drama") & 
            (mov$Studio=="Buena Vista Studios"|
               mov$Studio=="Fox"|
            mov$Studio=="Paramount Pictures"|
            mov$Studio=="Sony"|
            mov$Studio=="Universal"|
            mov$Studio=="WB"),]

windowsFonts(CS=windowsFont("Comic Sans MS"))

a<- ggplot(data=mov2, aes(x=Genre, y=GrosspercentUS))
b<-a+geom_jitter(aes(size=Budgetmillions, colour=Studio))+
  geom_boxplot(alpha=0.7, outlier.colour=NA) + labs(size="Budget $M")

#?labs
#?par
#?geom_jitter
#axeslabels
options(repr.plot.width=10, repr.plot.height=10)
b + ylab("Gross % US") +
  ggtitle("Domestic Gross % by Genre")+
  theme(axis.title.x=element_text(colour="Blue", size=30, family="CS"),
        axis.title.y=element_text(colour="Blue", size=30, family="CS"),
        axis.text.x=element_text(size=20, family="CS"),
        axis.text.y=element_text(size=20),
        
        legend.title=element_text(size=20, family="CS"),
        legend.text=element_text(size=10, family="CS"),
        
        plot.title=element_text(hjust=0.5,
                                size=20,
                                family = "CS"),
        #text=element_text(family="CS")
        )

#q$labels$size<- "Budget $M"
```

```r
mydata<- read.csv(file.choose())
library(ggplot2)
suppressWarnings(suppressMessages(library(ggplot2)))
ggplot(data=mydata[mydata$carat<2.5,],aes(x=carat,y=price,
colour=clarity)) + geom_point(alpha=0.1) + geom_smooth()
```

## Investigating if price of a diamond increases with diamond clarity holds true.

**Diamond clarity scale from highest to least clarity: IF,VVS1,VVS2,VS1,VS2,SI1,SI2,I1**

From a sample of 53 940 diamonds we can see that higher clarity diamonds tend to be priced higher than lower clarity damonds for the same carat weight. There is mispricing happening at certain carat weights were the trend lines of a higher clarity diamond cross the trend line of a lower clarity diamonds
